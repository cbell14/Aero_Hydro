```python
from math import *
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
N = 50                   # number of points
x_start, x_end = -2.0, 2.0 # x boundaries
y_start, y_end = -1.0, 1.0 # y boundaries

x = np.linspace(x_start,x_end,N)
y = np.linspace(y_start,y_end,N)

X, Y, = np.meshgrid(x,y)
```

### Source & Source

```python
strength_source = 5.0          # source strength
x_source, y_source = -1.0, 0.0 # location
strength_sink = -5.0                     # strength of the sink
x_sink, y_sink = 1.0, 0.0                # location of the sink

r_source = np.sqrt(((X - x_source)**2 + (Y - y_source)**2)) #distance of source
r_sink = np.sqrt(((X - x_sink)**2 + (Y - y_sink)**2)) #distance of sink

#more efficient velocity calc

vel_source = np.zeros([len(X),len(Y),2])
vel_sink = np.zeros_like(vel_source)
vel_combo = np.zeros_like(vel_source)

vel_source[:,:,0] = strength_source/(2.0*pi) * (X-x_source)/(r_source**2)
vel_source[:,:,1] = strength_source/(2.0*pi) * (Y-y_source)/(r_source**2)

vel_sink[:,:,0] = strength_sink/(2.0*pi) * (X-x_sink)/(r_sink**2)
vel_sink[:,:,1] = strength_sink/(2.0*pi) * (Y-y_sink)/(r_sink**2)

vel_combo[:,:,0] = vel_source[:,:,0] + vel_sink[:,:,0]
vel_combo[:,:,1] = vel_source[:,:,1] + vel_sink[:,:,1]
```

```python
# plotting the streamlines of source
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, vel_source[:,:,0], vel_source[:,:,1], density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
```

```python
# plots the streamlines of sink
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, vel_sink[:,:,0], vel_sink[:,:,1], density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_sink, y_sink, color='#CD2305', s=80, marker='o', linewidth=0);
```

```python
# plots the streamlines of the pair source/sink
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, vel_combo[:,:,0], vel_combo[:,:,1], density=2.0, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter([x_source, x_sink], [y_source, y_sink], 
            color='#CD2305', s=80, marker='o', linewidth=0);
```

$$u_r = \frac{\partial \Phi}{\partial r} $$

$$u_{\theta} = \frac{1}{r}\frac{\partial \Phi}{\partial \theta} $$

we know $u_{\theta}$ is zero so we don't need to integrate it. The radial component integrates like this

$$u_r \partial r = \partial \Phi$$

$$\frac{\sigma}{2 \pi r} \partial r = \partial \Phi$$

$$\Phi = \frac{\sigma}{2 \pi}ln\left|r\right|$$

now convert to cartesian coordinates yielding

$$\Phi = \frac{\sigma}{2 \pi}ln\left(\sqrt{x^2 + y^2}\right)$$

now find the velocity in cartesian coordinates using

$$u = \frac{\partial \Phi}{\partial x}$$

$$v = \frac{\partial \Phi}{\partial y}$$

$$u = \frac{\sigma}{2 \pi} \frac{x - x_{source}}{{\left(x - x_{source}\right)}^2
{ + \left(y - y_{source}\right)}^2} $$

$$v = \frac{\sigma}{2 \pi} \frac{y - y_{source}}{{\left(x - x_{source}\right)}^2
{ + \left(y - y_{source}\right)}^2} $$

We will use the same parameters as before with the source location at $\left(-1,0\right)$ and $\sigma = 5$

```python
# calculcate potential
phi_source = strength_source/(2*pi) * np.log(r_source)
phi_sink = strength_sink/(2*pi) * np.log(r_sink)
phi = phi_source + phi_sink
```

```python
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
phi = plt.contourf(X,Y,phi,alpha=0.5,levels = np.linspace(np.min(phi),np.max(phi),100)\
                   ,cmap = plt.cm.jet_r)
#cmap set up color map, cm color map, jet is default color scale, _r uses inverted scale
#so colors flipped but not values
plt.colorbar(phi, orientation='vertical', format="%.2f")
```
