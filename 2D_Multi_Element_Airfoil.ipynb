```python
import numpy as np
from matplotlib import pyplot as plt
import math
import csv
import pprint
```

```python
with open ('./resources/MainFoil_N=50.csv') as file_name:
    coords_main = np.genfromtxt(file_name, delimiter=',')


x_main = np.empty((len(coords_main),1), dtype=float)
y_main = np.empty((len(coords_main),1), dtype=float)
    
for i in range(0,len(coords_main)):
    x_main[i] = coords_main[i,0]
    y_main[i] = coords_main[i,1]

val_x, val_y = 0.1, 0.2
x_min_main, x_max_main = x_main.min(), x_main.max()
y_min_main, y_max_main = y_main.min(), y_main.max()
x_start_main, x_end_main = x_min_main-val_x*(x_max_main-x_min_main), x_max_main+val_x*(x_max_main-x_min_main)
y_start_main, y_end_main = y_min_main-val_y*(y_max_main-y_min_main), y_max_main+val_y*(y_max_main-y_min_main)

with open ('./resources/FlapFoil_N=50.csv') as file_name:
    coords_flap = np.genfromtxt(file_name, delimiter=',')
    
x_flap = np.empty((len(coords_flap),1), dtype=float)
y_flap = np.empty((len(coords_flap),1), dtype=float)

for i in range(0,len(coords_main)):
    x_flap[i] = coords_flap[i,0]
    y_flap[i] = coords_flap[i,1]

x_min_flap, x_max_flap = x_flap.min(), x_flap.max()
y_min_flap, y_max_flap = y_flap.min(), y_flap.max()
x_start_flap, x_end_flap = x_min_flap-val_x*(x_max_flap-x_min_flap), x_max_flap+val_x*(x_max_flap-x_min_flap)
y_start_flap, y_end_flap = y_min_flap-val_y*(y_max_flap-y_min_flap), y_max_flap+val_y*(y_max_flap-y_min_flap)


%matplotlib inline
size = 10
plt.figure(figsize=(20,6))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_start_main, x_end_flap)
#plt.ylim(y_start_main, y_end_flap)
plt.title('Main Airfoil')
plt.plot(x_main, y_main, color='k', linestyle='-',linewidth=2)
plt.plot(x_flap, y_flap, color='k', linestyle='-',linewidth=2);
```

```python
class Panel:
    """Contains information related to a Panel"""
    def __init__(self, xa, ya, xb, yb):
        """Initializes the panel
        
        Params:
        -------
        xa, ya  float, coordinates of the first end point of the panel
        xb, yb  float, coordinates of the second end point of the panel
        """
        
        self.xa, self.ya = xa, ya
        self.xb, self.yb = xb, yb
        
        self.xc, self.yc = (xa+xb)/2, (ya+yb)/2 #center point
        
        self.length = math.sqrt((xb-xa)**2 + (yb-ya)**2) #length of panel
        
        #panel orientation
        if xb-xa <= 0.0:
            self.beta = math.acos((yb-ya)/self.length)
        elif xb-xa> 0.0:
            self.beta = math.pi + math.acos(-(yb-ya)/self.length)
            
        #panel location
        if self.beta <= math.pi:
            self.loc = 'extrados'
        else:
            self.loc = 'intrados'
        
        self.sigma = 0.0   #source strength
        self.vt = 0.0      #tangential velocity
        self.cp = 0.0      #pressure coefficient
```

```python
N_main = len(x_main) -1 #number of panels
N_flap = len(x_flap) -1
panels_main = np.empty(N_main, dtype=object)
panels_flap = np.empty(N_flap, dtype=object)

for i in xrange(N_main):
    panels_main[i] = Panel(x_main[i], y_main[i], x_main[i+1], y_main[i+1])
    
for i in xrange(N_flap):
    panels_flap[i] = Panel(x_flap[i], y_flap[i], x_flap[i+1], y_flap[i+1])

%matplotlib inline
plt.figure(figsize=(20,6))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_start_main, x_end_flap)
#plt.ylim(y_start, y_end)
plt.plot(x_main, y_main, color='k', linestyle='-', linewidth=3)
plt.scatter([p.xa for p in panels_main],[p.ya for p in panels_main], color='r',\
    s=60)
plt.plot(x_flap, y_flap, color='k', linestyle='-',linewidth=3)
plt.scatter([p.xa for p in panels_flap],[p.ya for p in panels_flap], color='r',\
    s=60);
```

```python
class Freestream:
    """Freestream conditions"""
    def __init__(self, u_inf=1.0, alpha=0.0):
        """Sets the freestream conditions
        
        Params:
        -------
        u_inf    float, far field speed defaults to 1.0
        alpha    float, angle of attack in degrees defaults to 0.0
        """
        
        self.u_inf = u_inf
        self.alpha = alpha = alpha*math.pi/180
```
