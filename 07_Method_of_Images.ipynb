```python
import numpy as np
import math
from matplotlib import pyplot as plt
from pylab import rcParams
rcParams['font.size'] = 16
rcParams['figure.figsize'] = (20,16)
```

```python
N = 50
x_s, x_e = -2.0, 2.0
y_s, y_e = -1.0, 1.0
x = np.linspace(x_s,x_e,N)
y = np.linspace(y_s,y_e,N)
X, Y = np.meshgrid(x,y)
```

```python
class Source:
    """contains information about sources or sinks"""
    def __init__(self, strength, x, y):
        """Initialize the source or sink

        Params:
        -------
        strength  float, strength of the source or sink
        x, y      float, cartesian locations of the source or sink
        """

        self.strength = strength
        self.x, self.y = x, y

    def velocity(self, X, Y):
        """Computes the cartesian velocity field produced by
        a source or sink

        Params:
        -------
        X, Y     2D array of float, meshgrid
        """

        self.u = (self.strength/(2.0*math.pi)) * (X-self.x) /\
            ((X-self.x)**2 + (Y-self.y)**2)
        self.v = (self.strength/(2.0*math.pi)) * (Y-self.y) /\
            ((X-self.x)**2 + (Y-self.y)**2)

    def streamfunction(self, X, Y):
        """Computes the cartesian streamfunction produced by a
        source or sink

        Params:
        ------
        X, Y     2D array of float, meshgrid
        """

        self.psi = self.strength/(2.0*math.pi) * np.arctan2\
            ((Y-self.y),(X-self.x))
```

```python
strength_source = 1.0
x_source, y_source = 0.0, 0.5

#creates a source (the object of class source)
source = Source(strength_source, x_source, y_source)

#computes the velocity field and the stream-fucntion on the mesh grid
source.velocity(X,Y)
source.streamfunction(X,Y)
```

```python
#creates the image of the source and computes 
#velocity and streamfunction

source_image = Source(strength_source, x_source, -y_source)
source_image.velocity(X,Y)
source_image.streamfunction(X,Y)
```

```python
#superposition of the source and its image
u = source.u + source_image.u
v = source.v + source_image.v
psi = source.psi + source_image.psi

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_s,x_e)
plt.ylim(y_s,y_e)
plt.streamplot(X, Y, u, v, density=2.5, linewidth=1.5, arrowsize=1,\
    arrowstyle='->')
plt.scatter(source.x, source.y, color='r', s=250, marker='o')
plt.scatter(source_image.x, source_image.y, color='r', s=250,\
    marker='D')
plt.axhline(0.0, color='k', linestyle='--', linewidth=6);
```

```python
class Vortex:
    """Contains information about a vortex"""
    def __init__(self, strength, x, y):
        """Initialize the vortex

        Params:
        -------
        stength  float, strength of the vortex
        x, y     float, cartesian locations of the source or sink
        """

        self.strength = strength
        self.x, self.y = x, y

    def velocity(self, X, Y):
        """Computes the cartesian velocity field created by a vortex

        Params:
        -------
        X, Y     2D array of float, meshgrid
        """

        self.u = +self.strength/(2.0*math.pi) * (Y-self.y) / ((X-self.x)**2 + (Y-self.y)**2)
        self.v = -self.strength/(2.0*math.pi) * (X-self.x) / ((X-self.x)**2 + (Y-self.y)**2)

    def streamfunction(self, X, Y):
        """Computes the cartesian streamfunction created by a vortex

        Params:
        -------
        X, Y     2D array of float, meshgrid
        """

        self.psi = -self.strength/(4.0*math.pi) * np.log((X-self.x)**2 + (Y-self.y)**2)
```

```python
strength_vort = 1.0
x_vort, y_vort = 0.0, 0.5

vortex = Vortex(strength_vort, x_vort, y_vort)
vortex.velocity(X, Y)
vortex.streamfunction(X, Y)

vortex_image = Vortex(-strength_vort, x_vort, -y_vort)
vortex_image.velocity(X, Y)
vortex_image.streamfunction(X, Y)
```

```python
u = vortex.u + vortex_image.u
v = vortex.v + vortex_image.v
psi = vortex.psi + vortex_image.psi

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_s,x_e)
plt.ylim(y_s,y_e)
plt.streamplot(X, Y, u, v, density=2.5, linewidth=1.5, arrowsize=1,\
    arrowstyle='->')
plt.scatter(vortex.x, vortex.y, color='r', s=250, marker='o')
plt.scatter(vortex_image.x, vortex_image.y, color='r', s=250,\
    marker='D')
plt.axhline(0.0, color='k', linestyle='--', linewidth=6);
```

```python
strength_vort = 1.0
x_vort1, y_vort1 = -0.1, 0.5
x_vort2, y_vort2 = 0.1, 0.5

vortex1 = Vortex(strength_vort, x_vort1, y_vort1)
vortex2 = Vortex(-strength_vort, x_vort2, y_vort2)
vortex1_image = Vortex(-strength_vort, x_vort1, -y_vort1)
vortex2_image = Vortex(strength_vort, x_vort2, -y_vort2)

vortex1.velocity(X, Y)
vortex1.streamfunction(X, Y)
vortex2.velocity(X, Y)
vortex2.streamfunction(X, Y)

vortex1_image.velocity(X, Y)
vortex1_image.streamfunction(X, Y)
vortex2_image.velocity(X, Y)
vortex2_image.streamfunction(X, Y)
```

```python
u = vortex1.u + vortex2.u + vortex1_image.u + vortex2_image.u
v = vortex1.v + vortex2.v + vortex1_image.v + vortex2_image.v
psi = vortex1.psi + vortex2.psi + vortex1_image.psi\
    + vortex2_image.psi
    
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_s,x_e)
plt.ylim(y_s,y_e)
plt.streamplot(X, Y, u, v, density=2.5, linewidth=1.5, arrowsize=1,\
    arrowstyle='->')
plt.scatter(vortex1.x, vortex1.y, color='r', s=250, marker='o')
plt.scatter(vortex2.x, vortex2.y, color='g', s=250, marker='o')
plt.scatter(vortex1_image.x, vortex1_image.y, color='r', s=250,\
    marker='D')
plt.scatter(vortex2_image.x, vortex2_image.y, color='g', s=250,\
    marker='D')
plt.axhline(0.0, color='k', linestyle='--', linewidth=6)
plt.axvline(0.0, color='k', linestyle='--', linewidth=6);
```

```python
class Doublet:
    """Contains information about a doublet"""
    def __init__(self, strength, x, y):
        """Initialize the doublet

        Params:
        -------
        strength  float, strength of the doublet
        x, y      float, cartesian locations of the doublet
        """

        self.strength = strength
        self.x, self.y = x, y

    def velocity(self, X, Y):
        """Computes the cartesian velocity field created by a
        doublet

        Params:
        ------
        X, Y     2D array of float, meshgrid
        """

        self.u = -self.strength/(2.0*math.pi) * ((X-self.x)**2\
            - (Y-self.y)**2)/((X-self.x)**2 + (Y-self.y)**2)**2
        self.v = -self.strength/(2.0*math.pi) * (2.0 * (X-self.x)\
            *(Y-self.y))/((X-self.x)**2 + (Y-self.y)**2)**2

    def streamfunction(self, X, Y):
        """Computes the cartesian streamfunction created by a
        doublet

        Params:
        ------
        X, Y    2D array of float, meshgrid
        """

        self.psi = -self.strength/(2.0*math.pi) * (Y-self.y)\
            /((X-self.x)**2 + (Y-self.y)**2)
```

```python
strength_doub = 1.0
x_doub, y_doub = 0.0, 0.3

doublet = Doublet(strength_doub, x_doub, y_doub)
doublet.velocity(X, Y)
doublet.streamfunction(X, Y)

doublet_image = Doublet(strength_doub, x_doub, -y_doub)
doublet_image.velocity(X, Y)
doublet_image.streamfunction(X, Y)
```

```python
u_inf = 1.0    # free-stream speed

# calculates the velocity and stream-function of the
#free-stream flow
u_freestream = u_inf * numpy.ones((N, N), dtype=float)
v_freestream = numpy.zeros((N, N), dtype=float)
psi_freestream = u_inf * Y
```

```python
u = u_freestream + doublet.u + doublet_image.u
v = v_freestream + doublet.v + doublet_image.v
psi = psi_freestream + doublet.psi + doublet_image.psi

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_s,x_e)
plt.ylim(y_s,y_e)
plt.streamplot(X, Y, u, v, density=2.5, linewidth=1.5, arrowsize=1,\
    arrowstyle='->')
plt.scatter(doublet.x, doublet.y, color='r', s=250, marker='o')
plt.scatter(doublet_image.x, doublet_image.y, color='r', s=250,\
    marker='D')
plt.axhline(0.0, color='k', linestyle='--', linewidth=6);
```
