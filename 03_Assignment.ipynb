```python
from functions import gen_uform_grid
from functions import vel_uniform_flow_alpha, sf_uniform_flow_alpha
from functions import vel_source_sink, sf_source_sink
```

```python
sigma = np.loadtxt('Resources/NACA0012_sigma.txt')
x_load = np.loadtxt('Resources/NACA0012_x.txt')
y_load = np.loadtxt('Resources/NACA0012_y.txt')
```

```python
N = 51
x_start, x_end = -1.0,2.0
y_start, y_end = -0.5,0.5
y = np.linspace(-0.5,0.5,N)
u_inf = 1.0
alpha = 0.0
source_strength = sigma.copy()

X, Y = gen_uform_grid(N, x_start, x_end, y_start, y_end)
```

```python
def vel_combined_source_sink(strength, xs, ys, X, Y, N):
    """Calculates the velocities for several combined
    sources or sinks
    
    Params:
    ------
    strength   float, strength of sources/sinks
    xs, ys     1D array of float, coordinates of the sources/sinks
    X, Y       2D array of float, mesh grid
    N          float, number of points
    
    Returns:
    -------
    u_all, v_all      2D arrays of float, x and y velocities
    """
    
    #u_s, v_s = np.zeros((N,N)), np.zeros((N,N))
    u_all, v_all = np.zeros((N,N)), np.zeros((N,N))
    
    for i in range(0,len(source_strength)):
        u_s, v_s = vel_source_sink(strength[i], xs[i], ys[i], X, Y)
        u_all += u_s
        v_all += v_s
        
    return u_all, v_all
```

```python
def sf_combined_source_sink(strength, xs, ys, X, Y, N):
    """Calculates the stream functions for several combined
    sources or sinks
    
    Params:
    ------
    strength   float, strength of sources/sinks
    xs, ys     1D array of float, coordinates of the sources/sinks
    X, Y       2D array of float, mesh grid
    N          float, number of points
    
    Returns:
    -------
    psi_s        2D array of float, streamfunction
    """
    
    #psi_s = np.zeros((N,N))
    psi_all = np.zeros((N,N))
    
    for i in range(0,len(source_strength)):
        psi_s = sf_source_sink(strength[i], xs[i], ys[i], X, Y)
        psi_all += psi_s
        
    return psi_all
```

```python
u_free, v_free = vel_uniform_flow_alpha(u_inf, alpha, N, X, Y)
psi_free = sf_uniform_flow_alpha(u_inf, alpha, X, Y)
u_sources, v_sources = vel_combined_source_sink(source_strength,\
    x_load, y_load, X, Y, N)
psi_sources = sf_combined_source_sink(source_strength, x_load,\
    y_load, X, Y, N)
```

```python
u = u_free + u_sources
v = v_free + v_sources
psi = psi_free + psi_sources
```

```python
# computes the pressure coefficient field
cp = 1.0 - (u**2+v**2)/u_inf**2
cp_max = np.max(cp)
cp_max_loc = np.argmax(cp)
x_loc = X.flat[cp_max_loc] #argmin gives linear index so use flat to
#generate flat matrix to reference
y_loc = Y.flat[cp_max_loc]
print("The maximum C_p is %.4f at %.4d (linear index)" % (cp_max, cp_max_loc))
print("The stagnation point is located at %.4f,%.4f" % (x_loc, y_loc))

#using unravel
cp_max_loc2 = numpy.unravel_index(cp_max_loc, cp.shape)
print(cp_max_loc2)
```
