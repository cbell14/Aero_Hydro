```python
import math
import numpy as np
from scipy import integrate
from matplotlib import pyplot as plt
import pylab
rcParams['font.size'] = 16
rcParams['figure.figsize'] = (10,8)
```

```python
with open ('./resources/naca0012.dat') as file_name:
    x, y = np.loadtxt(file_name, dtype=float, delimiter='\t', unpack=True)

val_x, val_y = 0.1, 0.2
x_min, x_max = x.min(), x.max()
y_min, y_max = y.min(), y.max()
x_start, x_end = x_min-val_x*(x_max-x_min), x_max+val_x*(x_max-x_min)
y_start, y_end = y_min-val_y*(y_max-y_min), y_max+val_y*(y_max-y_min)

size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.plot(x, y, color='k', linestyle='-',linewidth=2);
```

```python
class Panel:
    """Contains information related to a Panel"""
    def __init__(self, xa, ya, xb, yb):
        """Initializes the panel
        
        Params:
        -------
        xa, ya  float, coordinates of the first end point of the panel
        xb, yb  float, coordinates of the second end point of the panel
        """
        
        self.xa, self.ya = xa, ya
        self.xb, self.yb = xb, yb
        
        self.xc, self.yc = (xa+xb)/2, (ya+yb)/2 #center point
        
        self.length = math.sqrt((xb-xa)**2 + (yb-ya)**2) #length of panel
        
        #panel orientation
        if xb-xa <= 0.0:
            self.beta = math.acos((yb-ya)/self.length)
        elif xb-xa> 0.0:
            self.beta = math.pi + math.acos(-(yb-ya)/self.length)
            
        #panel location
        if self.beta <= math.pi:
            self.loc = 'extrados'
        else:
            self.loc = 'intrados'
        
        self.sigma = 0.0   #source strength
        self.vt = 0.0      #tangential velocity
        self.cp = 0.0      #pressure coefficient
```

```python
def define_panels(x, y, N=40):
    """Discretizes the geometry into panels using 'cosine' method
    
    Params:
    -------
    x, y     1D array of float, cartesian coordinates of the geometry
    N        float, number of panels defaults to 40
    """
    
    R = (x.max()-x.min())/2  #radius of the circle
    x_center = (x.max()+x.min())/2 #x-coord of the center
    x_circle = x_center + R*np.cos(np.linspace(0,2*math.pi,N+1)) #x-coords
    #of the circle points
    
    x_ends = np.copy(x_circle) #projection of circle x-coords onto surface
    y_ends = np.empty_like(x_ends)
    
    x, y = np.append(x, x[0]), np.append(y, y[0]) #extending arrays by 1
    
    #interpolating y-coordinates of end points
    I = 0
    for i in range(N):
        while I < len(x) - 1:
            if (x[I] <= x_ends[i] <= x[I+1]) or (x[I+1]\
                <= x_ends[i] <= x[I]):
                break
            else:
                I += 1
        a = (y[I+1]-y[I])/(x[I+1]-x[I])
        b = y[I+1] - a*x[I+1]
        y_ends[i] = a*x_ends[i] + b
    y_ends[N] = y_ends[0] # first and last points must be the same to meet
    
    panels = np.empty(N, dtype=object)
    for i in range(N):
        panels[i] = Panel(x_ends[i], y_ends[i], x_ends[i+1], y_ends[i+1])
        
    return panels
```

```python
N = 40  #number of panels
panels = define_panels(x, y, N)

val_x, val_y = 0.1, 0.2
x_min, x_max = min(panel.xa for panel in panels), max(panel.xa for panel\
    in panels)
y_min, y_max = min(panel.ya for panel in panels), max(panel.ya for panel\
    in panels)
x_start, x_end = x_min-val_x*(x_max-x_min), x_max+val_x*(x_max-x_min)
y_start, y_end = y_min-val_y*(y_max-y_min), y_max+val_y*(y_max-y_min)

size=10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.plot(x, y, color='k', linestyle='-', linewidth=3)
plt.plot(np.append([panel.xa for panel in panels], panels[0].xa),\
    np.append([panel.ya for panel in panels], panels[0].ya),\
    color='r', linestyle='-', linewidth=1, marker='o', markersize=6);
```

```python
class Freestream:
    """Freestream conditions"""
    def __init__(self, u_inf=1.0, alpha=0.0):
        """Sets the freestream conditions
        
        Params:
        -------
        u_inf    float, far field speed defaults to 1.0
        alpha    float, angle of attack in degrees defaults to 0.0
        """
        
        self.u_inf = u_inf
        self.alpha = alpha = alpha*math.pi/180
```

```python
u_inf, alpha = 1.0, 0.0
freestream = Freestream(u_inf, alpha)
```

```python
def integral(x, y, panel, dxdz, dydz):
    """Evaluates the contribution of a panel at one point
    
    Parms:
    ------
    x, y    1D array of float, cartesian coordinates of point
    panel   object, panel which contribution is evaluated
    dxdz    derivative of x in z-direction
    dydz    derivatice of y in z direction
    
    Returns:
    -------
    Integral over the panel of the influence at one point
    """
    
    def func(s):
        return (((x - (panel.xa - math.sin(panel.beta)*s))*dxdz\
            +(y - (panel.ya + math.cos(panel.beta)*s))*dydz)\
            / ((x - (panel.xa - math.sin(panel.beta)*s))**2\
            +(y - (panel.ya + math.cos(panel.beta)*s))**2))
    return integrate.quad(lambda s:func(s), 0.0, panel.length)[0]
```

```python
def build_matrix(panels):
    """Builds the source matrix
    
    Params:
    ------
    panels   object, array of panels
    
    Returns:
    -------
    A        2D array of float, NxN matrix
    """
    N = len(panels)
    A = np.empty((N,N), dtype=float)
    np.fill_diagonal(A, 0.5)
    
    for i, p_i in enumerate(panels):
        for j, p_j in enumerate(panels):
            if i != j:
                A[i,j] = 0.5/math.pi*integral(p_i.xc, p_i.yc, p_j,\
                    math.cos(p_i.beta), math.sin(p_i.beta))
                
    return A

def build_RHS(panels, freestream):
    """Builds the RHS of the linear system
    
    Params:
    ------
    panels      object, array of panels
    freestream  object, freestream conditions
    
    Returns:
    -------
    b           1D array of float, (N+1)x1 matrix
    """
    
    b = np.empty(len(panels), dtype=float)
    
    for i, panel in enumerate(panels):
        b[i] = -freestream.u_inf * math.cos(freestream.alpha - panel.beta)
        
    return b
```

```python
A = build_matrix(panels)
b = build_RHS(panels, freestream)

#solve for sigma
sigma = np.linalg.solve(A, b)

#assign sigma values to panels
for i, panel in enumerate(panels):
    panel.sigma = sigma[i]
```

```python
def get_tangential_velocity(panels, freestream):
    """Computes the tangential velocity on the surface
    
    Params:
    -------
    panels     object, array of panels
    freestream object, freestream conditions
    """
    
    N = len(panels)
    A = np.empty((N,N), dtype=float)
    np.fill_diagonal(A, 0.0)
    
    for i, p_i in enumerate(panels):
        for j, p_j in enumerate(panels):
            if i != j:
                A[i,j] = 0.5/math.pi*integral(p_i.xc, p_i.yc, p_j,\
                    -math.sin(p_i.beta), math.cos(p_i.beta))
    
    b = freestream.u_inf * np.sin([freestream.alpha - panel.beta\
        for panel in panels])
    
    sigma = np.array([panel.sigma for panel in panels])
    
    vt = np.dot(A, sigma) + b  #calculating vt
    
    for i, panel in enumerate(panels):  #assigning vt to each panel
        panel.vt = vt[i]
        
def get_pressure_coefficient(panels, freestream):
    """Computes the surface pressure coefficients
    
    Parmas:
    -------
    panels     object, array of panels
    freestream object, freestream conditions
    """
    
    for panel in panels:
        panel.cp = 1.0 - (panel.vt/freestream.u_inf)**2
```

```python
get_tangential_velocity(panels, freestream)
get_pressure_coefficient(panels, freestream)
```

```python
x_min, x_max = min(panel.xa for panel in panels), max(panel.xa for\
    panel in panels)
cp_min, cp_max = min(panel.cp for panel in panels), max(panel.cp\
    for panel in panels)
x_start, x_end = x_min-val_x*(x_max-x_min), x_max+val_x*\
    (x_max-x_min)
y_start, y_end = cp_min-val_y*(cp_max-cp_min), cp_max+val_y*\
    (cp_max-cp_min)
    
plt.figure(figsize=(10,6))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('$C_p$')
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.title('Number of panels: %d' % N)
plt.plot([panel.xc for panel in panels if panel.loc == 'extrados'],\
    [panel.cp for panel in panels if panel.loc == 'extrados'],\
    color='r', linestyle='-', linewidth=2, marker='o', markersize=6)
plt.plot([panel.xc for panel in panels if panel.loc == 'intrados'],\
    [panel.cp for panel in panels if panel.loc == 'intrados'],\
    color='b', linestyle='-', linewidth=1, marker='o', markersize=6)
plt.legend(['extrados', 'intrados'], loc='best', prop={'size':14})
plt.gca().invert_yaxis();
```

```python
def get_velocity_field(panels, freestream, X, Y):
    """Calculates the velocity field
    
    Params:
    -------
    panels     object, array of panels
    freestream object, freestream conditions
    X, Y       2D array of float, mesh grid
    """
    
    Nx, Ny = X.shape
    u, v = np.empty((Nx, Ny), dtype=float), np.empty((Nx, Ny),\
        dtype=float)
    
    for i in xrange(Nx):
        for j in xrange(Ny):
            u[i,j] = freestream.u_inf*math.cos(freestream.alpha)\
                + 0.5/math.pi*sum([p.sigma * integral(X[i,j],\
                Y[i,j], p, 1, 0) for p in panels])
                
            v[i,j] = freestream.u_inf*math.sin(freestream.alpha)\
                + 0.5/math.pi*sum([p.sigma * integral(X[i,j],\
                Y[i,j], p, 0, 1) for p in panels])
                
    return u, v
```

```python
Nx, Ny = 20, 20
val_x, val_y = 1.0, 2.0
x_min, x_max = min(panel.xa for panel in panels), max(panel.xa for\
    panel in panels)
y_min, y_max = min(panel.ya for panel in panels), max(panel.ya for\
    panel in panels)
x_start, x_end = x_min-val_x*(x_max-x_min), x_max+val_x*\
    (x_max-x_min)
y_start, y_end = y_min-val_y*(y_max-y_min), y_max+val_y*\
    (y_max-y_min)
    
X, Y = np.meshgrid(np.linspace(x_start,x_end,Nx), np.linspace(\
    y_start,y_end,Ny))

u, v = get_velocity_field(panels, freestream, X, Y)
```

```python
size=10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.title('Streamlines around a NACA 0012 airfoil, AoA = %.1f'\
    %alpha)
plt.streamplot(X, Y, u, v, density=1, linewidth=1, arrowsize=1,\
    arrowstyle='->')
plt.fill([panel.xc for panel in panels], [panel.yc for panel in\
    panels], color='k', linestyle='solid', linewidth=2, zorder=2);
```
