```python
import math
import numpy as np
from matplotlib import pyplot as plt
from functions import gen_uform_grid, vel_doublet, sf_doublet,\
    vel_uniform_flow_alpha, sf_uniform_flow_alpha,\
    vel_vortex, sf_vortex
from pylab import rcParams
rcParams['font.size'] = 16
rcParams['figure.figsize'] = (20.0,16.0)
```

```python
N = 100
x_start, x_end = -2.0, 2.0
y_start, y_end = -1.0, 1.0
alpha = 0.0
u_inf = 1.0
kappa = 1.0
x_doub, y_doub = 0.0, 0.0 #location of doublet

X, Y = gen_uform_grid(N, x_start, x_end, y_start, y_end)
u_doub, v_doub = vel_doublet(kappa, x_doub, y_doub, X, Y)
psi_doub = sf_doublet(kappa, x_doub, y_doub, X, Y)
u_free, v_free = vel_uniform_flow_alpha(u_inf, alpha, N, X, Y)
psi_free = sf_uniform_flow_alpha(u_inf, alpha, X, Y)
```

```python
#superposition
u = u_doub + u_free
v = v_doub + v_free
psi = psi_doub + psi_free
```

```python
size = 16
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u, v, density=2.0, linewidth=1.0, arrowsize=1,\
    arrowstyle='->')
plt.scatter(x_doub, y_doub, color='r', s=120, marker='o')

R = math.sqrt(kappa/(2.0*math.pi*u_inf))
circle = plt.Circle((0,0), radius=R, color='r', alpha=0.5)
plt.gca().add_patch(circle)

x_stagn1, y_stagn1 = math.sqrt(kappa/(2.0*math.pi*u_inf)), 0
x_stagn2, y_stagn2 = -math.sqrt(kappa/(2.0*math.pi*u_inf)), 0
plt.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g',\
    s=120, marker ='o')
```

```python
#adding vortex
R = math.sqrt(kappa/(2.0*math.pi*u_inf))
#gamma = 4.0
gamma = 4.0 * math.pi * u_inf * R
x_vort, y_vort = 0.0, 0.0

u_vort, v_vort = vel_vortex(gamma, x_vort, y_vort, X, Y)
psi_vort = sf_vortex(gamma, x_vort, y_vort, X, Y)
```

```python
#super position
u = u_vort + u_doub + u_free
v = v_vort + v_doub + v_free
psi = psi_vort + psi_doub + psi_free
```

```python
x_stagn1, y_stagn1 = math.sqrt(R**2-(gamma/(4*math.pi*u_inf))**2),\
    -gamma/(4*math.pi*u_inf)
x_stagn2, y_stagn2 = -math.sqrt(R**2-(gamma/(4*math.pi*u_inf))**2),\
    -gamma/(4*math.pi*u_inf)
    
size = 16
plt.figure(figsize=(size,(y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1,\
    arrowstyle='->')
circle = plt.Circle((0,0), radius=R, color='r', alpha = 0.5)
pyplot.gca().add_patch(circle)
plt.scatter(x_vort, y_vort, color='r', s=120, marker='o')
plt.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g',\
    s=120, marker='o')
```

Challenge task

The challenge task in the doublet notebook was to calculate the radius of the cylinder created by the doublet in a uniform flow. You should have gotten

$$R=\sqrt{\frac{\kappa}{2\pi U_{\infty}}}$$
The new challenge is to find where the stagnation points are located on the surface of the cylinder, when there's a vortex. (You just need an expression for the angles.)

What hapens if  $\frac{\Gamma}{4\pi U_{\infty} R} > 1$?

Go back and experiment with a value of Γ that causes this

As gamma is changed the stagnation points move. When the expression equals 1 the stagnation points come together below the cylinder. When the expression is greater than 1 the expression moves off and below the cylinder

```python
theta = np.linspace(0 , 2*math.pi, 100)
u_theta = -2.0*u_inf*np.sin(theta) - gamma/(2*math.pi*R)

cp = 1.0 - (u_theta/u_inf)**2

u_theta_no_vort = -2.0*u_inf*np.sin(theta)
cp_no_vort = 1.0 - (u_theta_no_vort/u_inf)**2

size = 8
plt.figure(figsize=(size,size))
plt.grid(True)
plt.xlabel('$\theta$')
plt.ylabel('$C_p$')
plt.xlim(theta.min(), theta.max())
plt.plot(theta, cp, color='r', linewidth=2, linestyle='-')
plt.plot(theta, cp_no_vort, color='g', linewidth=2, linestyle='-')
plt.legend(['with vortex', 'without vortex'], loc='best', prop=\
    {'size':16})
```

Lift and Drag\
The lift is the component of force perpendicular to U∞, while the drag is the component parallel to U∞. How could we get them with the information we have above?

Well, the force on the cylinder is a product of the pressure acting on its surface (there is no viscosity here: it's ideal flow). If you draw a free body diagram, you should see that:

$$D = -\int_0^{2\pi} p \cos\left(\theta\right) R d\theta$$
$$L = -\int_0^{2\pi} p \sin\left(\theta\right) R d\theta$$

Using Bernoulli's equation, replace p in the equations above to obtain the lift and drag.

What does this mean?

We know that
$$p = p_{\infty} + \frac{1}{2} \rho\left[u_{\infty}^2 - \left(2 u_{\infty} \sin\theta - \frac{\Gamma}{2\pi R} \right)^2 \right]$$

For lift and drag this yields

$$L = -R\int_0^{2\pi} \left(p_{\infty} + \frac{1}{2} \rho\left[u_{\infty}^2 - \left(2 u_{\infty} \sin\theta - \frac{\Gamma}{2\pi R} \right)^2 \right] \right) \sin\theta d\theta$$

$$D = -R\int_0^{2\pi} \left(p_{\infty} + \frac{1}{2} \rho\left[u_{\infty}^2 - \left(2 u_{\infty} \sin\theta - \frac{\Gamma}{2\pi R} \right)^2 \right] \right) \cos\theta d\theta$$

Using some tricks from MIT's Aerodyamics class we can assign:
$$ A = \left(p_{\infty} + \frac{1}{2}\rho u_{\infty}^2 - \frac{\rho \Gamma^2}{2\pi^2 R^2}\right)$$

$$ B = \frac{\rho u_{\infty}\Gamma}{\pi R}$$

$$ C = -2\rho u_{\infty}^2$$

This simplfies Lift and Drag to 

$$L = -R \int_0^{2\pi} \left(A\sin\theta + B\sin^2\theta + C\sin^3\theta\right)d\theta$$

$$ D = -R \int_0^{2\pi} \left(A\sin\theta + B\sin\theta\cos\theta + C\sin^2\theta\cos\theta\right)d\theta$$

For Lift this yields an answer of $L = -BR\pi$ after integration

Substituting in for B
$$L = -\frac{\rho u_{\infty} \Gamma}{\pi R} R \pi$$
$$ L = -\rho u_{\infty}\Gamma$$

For Drag after performing the integration we get
$$D = 0$$
This is D'Alembert's paradox because we have an inviscid solution
