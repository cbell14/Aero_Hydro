```python
class Panel:
    def __init__(self, xa, ya, xb, yb):
        self.xa, self.ya = xa, ya
        self.xb, self.yb = xb, yb
        self.xc, self.yc = (xa+xb)/2, (ya+yb)/2
        self.length = np.sqrt((xb-xa)**2+(yb-ya)**2)
        if xb-xa <= 0.:
            self.beta = np.arccos((yb-ya)/self.length)
        elif xb-xa > 0.:
            self.beta = np.pi + np.arccos(-(yb-ya)/self.length)

        # project to normal and tangential direction
        self.nx = np.cos(self.beta)
        self.ny = np.sin(self.beta)
        self.tx = -np.sin(self.beta)
        self.ty = np.cos(self.beta)

        if self.beta <= np.pi:
            self.loc = 'extrados'
        else:
            self.loc = 'intrados'
        
        self.sigma = 0.
        self.vt = 0.
        self.cp = 0.
```

```python
N_main = len(x_main) -1 #number of panels
N_flap = len(x_flap) -1
panels_main = np.empty(N_main, dtype=object)
panels_flap = np.empty(N_flap, dtype=object)

for i in xrange(N_main):
    panels_main[i] = Panel(x_main[i], y_main[i], x_main[i+1], y_main[i+1])
    
for i in xrange(N_flap):
    panels_flap[i] = Panel(x_flap[i], y_flap[i], x_flap[i+1], y_flap[i+1])

tot_panels = np.concatenate((panels_main,panels_flap))
%matplotlib inline
plt.figure(figsize=(20,6))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_start_main, x_end_flap)
#plt.ylim(y_start, y_end)
plt.plot(x_main, y_main, color='k', linestyle='-', linewidth=3)
plt.scatter([p.xa for p in panels_main],[p.ya for p in panels_main], color='r',\
    s=60)
plt.plot(x_flap, y_flap, color='k', linestyle='-',linewidth=3)
plt.scatter([p.xa for p in panels_flap],[p.ya for p in panels_flap], color='r',\
    s=60);
```

```python
class Freestream:
    """Freestream conditions"""
    def __init__(self, u_inf=1.0, alpha=0.0):
        """Sets the freestream conditions
        
        Params:
        -------
        u_inf    float, far field speed defaults to 1.0
        alpha    float, angle of attack in degrees defaults to 0.0
        """
        
        self.u_inf = u_inf
        self.alpha = alpha = alpha*math.pi/180
```

```python
u_inf, alpha = 1.0, 0.0
freestream = Freestream(u_inf, alpha)
```

```python
def integral(p_i, panel, a, b):
    """Evaluates the contribution of a panel at one point.
    
    Arguments
    ---------
    p_i -- the target panel
    x, y -- Cartesian coordinates of the point.
    panel -- panel which contribution is evaluated.
    
    Returns
    -------
    Integral over the panel of the influence at one point.
    """
    def f(s):
        return ( ((p_i.xc - (panel.xa - np.sin(panel.beta)*s))*a 
                  + (p_i.yc - (panel.ya + np.cos(panel.beta)*s))*b)
                / ((p_i.xc - (panel.xa - np.sin(panel.beta)*s))**2 
                   + (p_i.yc - (panel.ya + np.cos(panel.beta)*s))**2) )
    return integrate.quad(lambda s:f(s), 0., panel.length)[0]
```

```python
def source_matrix(panels):
    """Builds the source matrix
    
    Params:
    ------
    panels   object, array of panels
    
    Returns:
    ------
    A        2D array of float, NxN matrix
    """
    
    N=len(panels)
    A = np.empty((N, N), dtype=float)
    np.fill_diagonal(A, 0.5)
    
    for i, p_i in enumerate(panels):
        for j, p_j in enumerate(panels):
            if i != j:
                A[i,j] = 0.5/np.pi*integral(p_i, p_j, p_i.nx,\
                    p_i.ny)
                
    return A
```

```python
def vortex_array(panels):
    """Build the vortex array
    
    Params:
    -------
    panels     object, array of panels
    
    Returns:
    -------
    a          1D array of float, Nx1 matrix
    """
    N = len(panels)
    main = panels[:N/2]
    flap = panels[N/2:]
    a = np.zeros((N, 2), dtype=float)
    for i, p_i in enumerate(panels):
        #main
        for j, p_j in enumerate(main):
            if i != j:
                a[i, 0] -= 0.5/np.pi*integral(p_i, p_j, p_i.ny, p_i.nx)
                
        for j, p_j in enumerate(flap):
            if i != (j+N/2):
                a[i, 1] -= 0.5/np.pi*integral(p_i, p_j, p_i.ny, p_i.nx)
                
    return a
```

```python
def kutta_array(panels, a, b):
    """Builds the Kutta-condition array.
    
    Arguments
    ---------
    panels -- array of panels.
    
    Returns
    -------
    a -- 1D array (Nx1, N is the number of panels).
    """
    p_a, p_b = panels[a], panels[b]
    N = len(panels)
    main = panels[:N/2]
    flap = panels[N/2:]
    A = np.zeros(N+2, dtype=float)
    
    # coefficient of source strength
    for j, p_j in enumerate(panels):
        if j != a:
            A[j] += 0.5/np.pi * integral(p_a, p_j, p_a.tx, p_a.ty)
        if j != b:
            A[j] += 0.5/np.pi * integral(p_b, p_j, p_b.tx, p_b.ty)
    
    # coefficient of vortex strength
    # if a,b on main airfoil
    if b <= N/2-1:
        # main airfoil
        for j, p_j in enumerate(main):
            if j != a:
                A[-2] -= 0.5/np.pi * integral(p_a, p_j, p_a.ty, -p_a.tx)
            else:
                A[-2] -= 0.5
            if j != b:
                A[-2] -= 0.5/np.pi * integral(p_b, p_j, p_b.ty, -p_b.tx)
            else:
                A[-2] -= 0.5
        # flap airfoil
        for j, p_j in enumerate(flap):
            A[-1] -= 0.5/np.pi * integral(p_a, p_j, p_a.ty, -p_a.tx)
            A[-1] -= 0.5/np.pi * integral(p_b, p_j, p_b.ty, -p_b.tx)
    
    # if a,b on flap airfoil
    elif a >= N/2:
        # main airfoil
        for j, p_j in enumerate(main):
            A[-2] -= 0.5/np.pi * integral(p_a, p_j, p_a.ty, -p_a.tx)
            A[-2] -= 0.5/np.pi * integral(p_b, p_j, p_b.ty, -p_b.tx)
        # flap airfoil
        for j, p_j in enumerate(flap):
            if (j+N/2) != a:
                A[-1] -= 0.5/np.pi * integral(p_a, p_j, p_a.ty, -p_a.tx)
            else:
                A[-1] -= 0.5
            if (j+N/2) !=b :
                A[-1] -= 0.5/np.pi * integral(p_b, p_j, p_b.ty, -p_b.tx)
            else:
                A[-1] -= 0.5
    
    return A
```

```python
def build_matrix(panels):
    N = panels.size
    A = np.empty((N+2, N+2), dtype=float)
    AS = source_matrix(panels)
    av = vortex_array(panels)
    k1 = kutta_array(panels, 0, N/2-1)
    k2 = kutta_array(panels, N/2, N-1)
    
    A[:N, :N] = AS[:,:]
    A[:N, N:] = av[:,:]
    A[-2, :] = k1[:]
    A[-1, :] = k2[:]
    
    return A
```

```python
A = build_matrix(tot_panels)
```

```python
def build_rhs(panels, freestream):
    N = panels.size
    b = np.zeros(N+2, dtype=float)
    for i, p_i in enumerate(panels):
        b[i] = - freestream.u_inf * np.cos(freestream.alpha - p_i.beta)
    # kutta condition main airfoil    
    b[-2] = -freestream.u_inf * (np.sin(freestream.alpha-panels[0].beta)\
        + np.sin(freestream.alpha-panels[N/2-1].beta))
    # kutta condition flap airfoil
    b[-1] = -freestream.u_inf * (np.sin(freestream.alpha-panels[N/2].beta)\
        + np.sin(freestream.alpha-panels[N-1].beta))
    return b
```

```python
b = build_rhs(tot_panels, freestream)
sigma = np.linalg.solve(A,b)

for i, panel in enumerate(tot_panels):
    panel.sigma = sigma[i]
gamma_main, gamma_flap = sigma[-2], sigma[-1]
```

```python
def get_tangential_velocity(panels, freestream, sigma):
    """Calculate the tangential velocity for each panel"""
    N = len(panels)
    main = panels[:N/2]
    flap = panels[N/2:]
    A = np.zeros((N, N+2), dtype=float)
    b = np.zeros(N, dtype=float)
    for i, p_i in enumerate(panels):
        # freestream
        b[i] = freestream.u_inf * np.sin(freestream.alpha-p_i.beta)
        # source
        for j, p_j in enumerate(panels):
            if i != j:
                A[i, j] = 0.5/np.pi *integral(p_i, p_j, p_i.tx, p_i.ty)
        # vortex
        # i on main
        if i < N/2:
            # main wing contribution
            for j, p_j in enumerate(main):
                if i != j:
                    A[i, -2] -= 0.5/np.pi * integral(p_i, p_j, p_i.ty,\
                        -p_i.tx)
                else:
                    A[i, -2] -= 0.5
            # flap wing contribution
            for j, p_j in enumerate(flap):
                A[i, -1] -= 0.5/np.pi * integral(p_i, p_j, p_i.ty,\
                    -p_i.tx)
        # i on flap
        else:
            # main wing contribution
            for j, p_j in enumerate(main):
                A[i,-2] -= 0.5/np.pi * integral(p_i, p_j, p_i.ty, -p_i.tx)
            # flap contribution
            for j, p_j in enumerate(flap):
                if i != (j+N/2):
                    A[i,-1] -= 0.5/np.pi * integral(p_i, p_j, p_i.ty,\
                        -p_i.tx)
                else:
                    A[i,-1] -= 0.5
    
    vt = np.dot(A, sigma) + b
    for i, panel in enumerate(panels):
        panel.vt = vt[i]
```

```python
def get_cp(panels, freestream):
    for panel in panels:
        panel.cp = 1 - (panel.vt/freestream.u_inf)**2
```

```python
def Get_L_D(panels):
    L = 0.0
    D = 0.0
    for panel in panels:
        L -= panel.cp*panel.length*panel.ny
        D += panel.cp*panel.length*panel.nx
    return L, D
```
