```python
import math
import numpy as np
from scipy import integrate
from matplotlib import pyplot as plt
import pylab
rcParams['font.size'] = 16
rcParams['figure.figsize'] = (10,8)
```

```python
u_inf = 1.0      #freestream speed

#geometry definition
R = 1.0
theta = np.linspace(0,2*math.pi, 100)
x_cylinder, y_cylinder = R*np.cos(theta), R*np.sin(theta) #cylinder coords

size = 6
plt.figure(figsize=(size,size))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-R-0.1,R+0.1)
plt.ylim(-R-0.1,R+0.1)
plt.plot(x_cylinder, y_cylinder, color ='b', linestyle='-', linewidth=3);
```

```python
class Panel:
    """Contains information related to a Panel"""
    def __init__(self, xa, ya, xb, yb):
        """Initializes the panel
        
        Params:
        -------
        xa, ya  float, coordinates of the first end point of the panel
        xb, yb  float, coordinates of the second end point of the panel
        """
        
        self.xa, self.ya = xa, ya
        self.xb, self.yb = xb, yb
        
        self.xc, self.yc = (xa+xb)/2, (ya+yb)/2 #center point
        
        self.length = math.sqrt((xb-xa)**2 + (yb-ya)**2) #length of panel
        
        #panel orientation
        if xb-xa <= 0.0:
            self.beta = math.acos((yb-ya)/self.length)
        elif xb-xa> 0.0:
            self.beta = math.pi + math.acos(-(yb-ya)/self.length)
        
        self.sigma = 0.0   #source strength
        self.vt = 0.0      #tangential velocity
        self.cp = 0.0      #pressure coefficient
```

```python
N_panels = 14  #number of panels

#defining end points of panels
x_ends = R*np.cos(np.linspace(0.0, 2*math.pi, N_panels+1))
y_ends = R*np.sin(np.linspace(0.0, 2*math.pi, N_panels+1))

panels = np.empty(N_panels, dtype=object)
for i in xrange(N_panels):
    panels[i] = Panel(x_ends[i], y_ends[i], x_ends[i+1], y_ends[i+1])
    
size = 8
plt.figure(figsize=(size,size))
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-R-0.1,R+0.1)
plt.ylim(-R-0.1,R+0.1)
plt.plot(x_cylinder, y_cylinder, color='b', linestyle='-', linewidth=1)
plt.plot(x_ends, y_ends, color='r', linestyle='-', linewidth=2)
plt.scatter([p.xa for p in panels],[p.ya for p in panels], color='r',\
    s=60)
plt.scatter([p.xc for p in panels],[p.yc for p in panels], color='k',\
    s=60, zorder=3)
plt.legend(['cylinder', 'panels', 'end-points', 'center points'],\
    loc='best', prop={'size':16});
```

```python
def integral_normal(p_i,p_j):
    """Evaluates the contribution of a panel at the center-point of
    another in the normal direction
    
    Params:
    ------
    p_i    float, panel on which the contribution is calculated
    p_j    float, panel form which the contribution is calculated
    
    Returns:
    -------
    Integral over the panel of influence at a control-point.
    """
    
    def func(s):
        return ((+(p_i.xc-(p_j.xa-math.sin(p_j.beta)*s))*math.cos(p_i\
            .beta) + (p_i.yc-(p_j.ya+math.cos(p_j.beta)*s))*math.sin(p_i\
            .beta)) / ((p_i.xc-(p_j.xa-math.sin(p_j.beta)*s))**2 + (p_i\
            .yc-(p_j.ya+math.cos(p_j.beta)*s))**2))
    
    return integrate.quad(lambda s:func(s), 0.0, p_j.length)[0]
```

```python
#Filling a matrix in A*sigma=b
A = np.empty((N_panels, N_panels), dtype=float)
np.fill_diagonal(A, 0.5)

for i, p_i in enumerate(panels):
    for j, p_j in enumerate(panels):
        if i !=j:
            A[i,j] = 0.5/math.pi*integral_normal(p_i,p_j)
            
#Filling b matrix
b = -u_inf * np.cos([p.beta for p in panels])

sigma = np.linalg.solve(A, b) #gets the sigma values

for i, panel in enumerate(panels):
    panel.sigma = sigma[i] #assigns sigma values to correspondign source
```

```python
def integral_tangential(p_i, p_j):
    """Evaluates the contribution of a panel at the center-point of
    another in the tangential direction
    
    Params:
    ------
    p_i     float, panel on which the contribution is calculated
    p_j     float, panel from which the contribution is calculated
    
    Returns:
    -------
    Integral over the panel of the influence at the control-point
    """
    
    def func(s):
        return ((-(p_i.xc-(p_j.xa-math.sin(p_j.beta)*s))*math.sin(p_i.beta\
            ) +(p_i.yc-(p_j.ya+math.cos(p_j.beta)*s))*math.cos(p_i.beta))\
            /((p_i.xc-(p_j.xa-math.sin(p_j.beta)*s))**2\
            + (p_i.yc-(p_j.ya+math.cos(p_j.beta)*s))**2))
    return integrate.quad(lambda s:func(s), 0.0, p_j.length)[0]
```

```python
#Filling A matrix for vt
A = np.empty((N_panels, N_panels), dtype=float)
np.fill_diagonal(A, 0.0)

for i, p_i in enumerate(panels):
    for j, p_j in enumerate(panels):
        if i !=j:
            A[i,j] = 0.5/math.pi*integral_tangential(p_i,p_j)
            
#filling b for vt
b = -u_inf * np.sin([panel.beta for panel in panels])

vt = np.dot(A, sigma) + b #calcualting all vt values

for i, panel in enumerate(panels):
    panel.vt = vt[i]  #assigning all vt values
```

```python
#calculating C_p
for panel in panels:
    panel.cp = 1.0 - (panel.vt/u_inf)**2
    
cp_exact = 1.0 - 4*(y_cylinder/R)**2
```

```python
plt.grid(True)
plt.xlabel('x')
plt.ylabel('$C_p$')
plt.xlim(-R,R)
plt.ylim(-4.0,2.0)
plt.plot(x_cylinder, cp_exact, color='b', linestyle='-', linewidth=1,\
    zorder=1)
plt.scatter([p.xc for p in panels], [p.cp for p in panels], color='r',\
    s=60, zorder=2)
plt.title('Number of panels: %d' % N_panels)
plt.legend(['analytical', 'source panel method'], loc='best', prop=\
    {'size':16});
```
