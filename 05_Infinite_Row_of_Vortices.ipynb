```python
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
from pylab import rcParams
#rcParams['figure.figsize'] = (20.0,16.0)
rcParams['font.size'] = 16
from functions import gen_uform_grid, vel_vortex, sf_vortex,\
    vel_uniform_flow_alpha, sf_uniform_flow_alpha
```

```python
N = 100 #number of pts each direction
xs, xe = -2.0, 2.0 #x start and end
ys, ye = -0.5, 0.5 #y start and end
n = 10 #number of vortices
alpha = 0.0
u_inf = 1.0

xv = np.linspace(xs,xe,n)
yv = np.zeros_like(xv)
gamma = 5.0
vort_strength = gamma * np.ones_like(xv)

X,Y = gen_uform_grid(N,xs,xe,ys,ye) #generate uniform grid
u_free, v_free = vel_uniform_flow_alpha(u_inf, alpha, N, X, Y)
psi_free = sf_uniform_flow_alpha(u_inf, alpha, X, Y)
```

```python
def vel_vort_combined(strength, xv, yv, X, Y, N):
    """Returns cartesian velocity field from vortex
    
    Params:
    ------
    strength   1D array of float, strength of vortices
    xv, yv     1D array of float, vortex coordinates
    X, Y       2D array of float, meshgrid
    N          float, number of points
    
    Returns
    -------
    u_all, v_all   2D array of float, x and y velocities
    """

    u_all, v_all = np.zeros((N,N)), np.zeros((N,N))
    
    for i in range(0,len(strength)):
        u_v, v_v = vel_vortex(strength[i], xv[i], yv[i], X, Y)
        u_all += u_v
        v_all += v_v
    
    return u_all, v_all

def sf_vort_combined(strength, xv, yv, X, Y, N):
    """Returns cartesian stream-function from vortex
    
    Params:
    ------
    strength   1D array of float, strength of vortex
    x_v, y_v   1D array of float, vortex coordinates
    X, Y       2D array of float, meshgrid
    N          float, number of points
    
    Returns
    -------
    psi         2D array of float, streamfunction
    """
    
    psi_v = np.zeros((N,N), dtype = float)
    psi_all = np.zeros((N,N), dtype = float)
    
    for i in range(0,len(strength)):
        psi_v = sf_vortex(strength[i], xv[i], yv[i], X, Y)
        psi_all += psi_v
    
    return psi_all
```

```python
#Finite row of vortices

u_vortices, v_vortices = vel_vort_combined(vort_strength,\
    xv, yv, X, Y, N)
u = u_vortices + u_free
v = v_vortices + v_free

psi_vortices = sf_vort_combined(vort_strength, xv, yv, X, Y, N)
psi = psi_vortices + psi_free
```

```python
plt.figure(figsize =(20, (ye-ys)/(xe-xs)*16))
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(xs,xe)
plt.ylim(ys,ye)
plt.streamplot(X, Y, u, v, density=2.0, linewidth=1,\
    arrowsize=1, arrowstyle='->')
plt.scatter(xv,yv, color = 'r', marker = 'o', s = 150)
plt.contour(X,Y, psi, levels = [0.], linewdith = 5, colors = 'r')
#%matplotlib inline
```

```python
#infinte row of vortices
#N = 100
x = np.linspace(xs,xe,N)
x_vortices = x[::N/(n-1)] #picking off x coordinates of vortex locations
y_vortices = np.zeros_like(x_vortices)
a = x_vortices[1] - x_vortices[0]
c = 2.0*math.pi/a

u_inf = ((gamma/(2.0*a)) * (np.sinh(c*Y)/(np.cosh(c*Y) - np.cos(c*X))))
v_inf = (-(gamma/(2.0*a)) * (np.sin(c*Y)/(np.cosh(c*Y) - np.cos(c*X))))
```

```python
plt.figure(figsize = (20, (ye-ys)/(xe-xs)*20))
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(xs,xe)
plt.ylim(ys,ye)
plt.streamplot(X, Y, u_inf, v_inf, density=1.0, linewidth=1,\
    arrowsize=1, arrowstyle='->')
plt.scatter(x_vortices,y_vortices, color = 'r', marker = 'o', s = 150)
```
